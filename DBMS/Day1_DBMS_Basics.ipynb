**File System**

Problems with FileSystem

* Data redundancy and inconsistency  
* difficulty in accessing data :- slower retirval due to unstructured querying capabilities
* data isolation :- challenges in correlating data across seperate files leading to data isolation
* integrity problem :- risk of inadvertent data alterations or deletion creatinf intergrity problems
* atomicity problem :- potential for data inconsistency due to incomplete operations, leading to atomicity problems.
* concurrent access anomalies :- conflicts and inconsistencis from simlutaneous data access/modifications, causing concurrent access anomalies

**DBMS**

* Data Access :- structured querying capabilities allow for quicker data access
* DataIsolation :- facilitates data integration, reducing data isolation issues
* Data Integrity :- features to prevent unauthorized data alterations, maitaining integrity
* atomicity problem :- supports transaction properties like atomicity, ensuring operations are either completed fully or not at all
* concurrent access anomalies :- advanced concurrency controls to manage multiple users accessing the database simultaneously,reducing anomalies.

**instance and schemas**

* the collection of information stored in the database at a particular moment is called an instance of the database.  

* the overall design of the database is called the database schema

**view of database(data abstraction)**
* View level :- this is the pinnacle of data abstraction, displaying only a portion of the entire database focusing on the user interest areas. it can represent multiple views of the same data, allowing user to access information through various application from the database.
  
* Logical level/conceptual :- above the physical level, this level showcases data as entity sets and their relationships,detailing the types and connections between stored data in the database.

* Physical level :- the internal schema details data storagae and access on hardware featuring the lowest of data abstraction with complex structures, predominantly managed by the database by the database administrator.
