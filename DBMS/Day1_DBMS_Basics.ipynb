**instance and schemas**

* the collection of information stored in the database at a particular moment is called an instance of the database.  

* the overall design of the database is called the database schema

**view of database(data abstraction)**
* View level :- this is the pinnacle of data abstraction, displaying only a portion of the entire database focusing on the user interest areas. it can represent multiple views of the same data, allowing user to access information through various application from the database.
  
* Logical level/conceptual :- above the physical level, this level showcases data as entity sets and their relationships,detailing the types and connections between stored data in the database.

* Physical level :- the internal schema details data storagae and access on hardware featuring the lowest of data abstraction with complex structures, predominantly managed by the database by the database administrator.

**Types of DataBase**

* Commercial Database :- Predominantly used in the business sector to handle large volumes of transactions and customer data. A CRM system like salesforce which handles large volumes of customer data and transaction.
* Multimedia Database :- Stores data types such as images, audio, and video files facilitating the management and retireval of multimedia content. A digital asset management system like Adobe experience Manager that facilitates the storage and retireval of multimedia content
* Deductive Database :- utilizes logic programming to derive information from data stored in a database, allowing for more complex and analytical queries. A database using DataLog(A query language) which allows for complex logical queries and information derivation.
* Temporal Database :- keeps track of chaging data over time, allowing for queries concerning time-based data. A historical trading database in the financial sectro which keeps track of stock prices over time.
* Geological information system :- Stores, organizes and analyzes geographical data, aiding in spatial analysis and mapping projects.

**DataBase Administrator** :- 
Database administrator holds authority over data and the programs facilitating data access. Their role/fucntions are:

* Schema Definition :-
  DBA outline the original database schema. Achieved through writing defintions translated to permanent labels in the data dictionary by the DDL compiler.
* Storage structure and access method definition :-
  Responsible for forming appropriate storage structures and access methods, Achieved through writing defintion translated by the data storage and definiton language compiler.
* Schema and physical Organisation Modification :-
  involves altering the databse schema or physical storage organization. changes are implemented by writing definitons that modify the relevant internal system tables.
* Granting of Authorization for data Access :-
  DBA grants varies types of Data Access authorization to different database users.
* Integrity constraint specification :-
  DBA implements and maintains integrity constrains to ensure data accuracy and consistency.
