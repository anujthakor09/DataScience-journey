## ER Diagram

* developed by Dr. Peter Chen in 1976, this conceptual level method, grounded in real-world preceptions, Facilitates diagrammatic data representations, simplifying comprehension for non-technical Users.

* The E-R data model, central to database design, encapsulates entities and their attributes within an enterprise schema, serving as a clear, standardized tool for translating real-world enterprise interaction into a conceptual schema.

**Entity** 

* AN entity is a thing or an object in the real world that is distinguishable from other object based on the values of the attributes it possesses.

* An entity may be a **concrete**, such as a person or a book,or it may be abstract, such as a course, a course offering, or a flight  reservation.

**Types of Entity**

* Tangible :- Entities which physically exist in the real world. eg. - Car,Pen,etc;

* Intangible :- Entities which exist logically. E.g. - Account, Video..


**In** ER diagram we cannot represent an entity, as entity is an instance not schema, and ER diagram is designed to understand schema.

**Entity Set**

* collection of same type of entities that share the same properties or attriutes.

* In an ER diagram an entity set is represented by a rectangle.

* In relational model it is represented by a seperate table.

* **Attributes** :-

      > Attributes are the units defines and descibes properties and characteristics of entities.

      > Attributes are the descriptive properties possessed by each member of an entity set. for each attribute there is a set of permitted values called domain.
