## ER Diagram

* developed by Dr. Peter Chen in 1976, this conceptual level method, grounded in real-world preceptions, Facilitates diagrammatic data representations, simplifying comprehension for non-technical Users.

* The E-R data model, central to database design, encapsulates entities and their attributes within an enterprise schema, serving as a clear, standardized tool for translating real-world enterprise interaction into a conceptual schema.

## **Entity** 

* AN entity is a thing or an object in the real world that is distinguishable from other object based on the values of the attributes it possesses.

* An entity may be a **concrete**, such as a person or a book,or it may be abstract, such as a course, a course offering, or a flight  reservation.

**Types of Entity**

* Tangible :- Entities which physically exist in the real world. eg. - Car,Pen,etc;

* Intangible :- Entities which exist logically. E.g. - Account, Video..


**In** ER diagram we cannot represent an entity, as entity is an instance not schema, and ER diagram is designed to understand schema.

**Entity Set**

* collection of same type of entities that share the same properties or attriutes.

* In an ER diagram an entity set is represented by a rectangle.

* In relational model it is represented by a seperate table.

* **Attributes** :-

      > Attributes are the units defines and descibes properties and characteristics of entities.

      > Attributes are the descriptive properties possessed by each member of an entity set. for each attribute there is a set of permitted values called domain.



**Attributes Representation**

  * 

  * 

  * In an ER diagram attributes are represented by an ellipse or oval connected to rectangle

  * while the relational model they are represented by independent column

**Types of Attributes**

**On the basis of Values**
  * Single valued :-
      * attributes having single value at any instance of time for an entity
      * eg. Adhar no., DOB
      * represented by single oval.
   
   * MultiValued :-
      * Attributes which can have more than one value for an entity at same time
      *  It is represented by a double ellipse in an ER diagram and by an independent table in a relational model
      *  Seperate table for each multivalued attribute, by taking mva and pk of main table as fk in new table

**On the Basis Of Division**
   * Simple :-
       * Attributes which cannot be divided further into subparts
       * Eg Age.
         
   * Composite :-
        * Attributes which can be further divided into subparts, as simple attributes. A composite attribute is represented by an ellise connected to an ellipse and in a relational model by a seperate column

**On the Basis Of Storage**
   * Stored :-
       * Main Attributes whose value is permanently stored in DataBase.
       * Eg DOB.
         
   *  Derived :-
        * The value of these types of attributes can be derived from values of other attributes 
    

## **Relationships**

* Is an association between two or more entities of same or different entity set.
* In ER diagram we cannot represent individual relationships as it is an instance or data.
* In an ER diagram it is represented by a diamond, while in relational model sometimes through foreign key and other time by a seperate table.

**Components**

   * Name : every relation must have a unique name
   * Degree :
       * Means number of entities set(relationship/tables) associated in relationship table.
       * Most of the relationship sets in a data base system are binary
       * Occasionally however relationships sets involve more than two entity sets.
       * Logically we can associate any number of entity set in a relationship called relationship.
   * Structural constraints :
       * An E-R enterprise schema may define certain contraints to which the contents of a database must conform.
