## group BY

```python
# Average Python marks by Gender.
# Average Attendance by Course.
# Maximum Python marks by Course.
# Count students in each Course.

# Mini Challenge

# Create a summary table using groupby().
```

```python
df1 = df.groupby('Gender')['Python_Marks'].mean()
```

```python
df1
```

```python
df.groupby('Course')["Attendance (%)"].mean()
```

```python
df.groupby('Course')['Student_ID'].value_counts()
```

```python
summary = df.groupby('Course').agg({
    'Python_Marks': ['mean', 'max', 'min'],
    'NumPy_Marks': ['mean', 'max'],
    'Attendance (%)': 'mean',
    'Student_ID': 'count'
})

summary
```
