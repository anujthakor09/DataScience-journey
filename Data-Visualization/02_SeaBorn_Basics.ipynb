```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Distribution Plots

```python
df = sns.load_dataset('tips')
```

**Dist/Hist plot**

```python
# distplot/histplot are used in showcasing the distribution of the data 

sns.displot(df['total_bill'],kde = True)
```

**JointPlot**

* used for Data comparision of two quant. entity attributes

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = df,kind='reg')
```

**PairPlot**

* plots and compare and deduce key points by comparing each numerical column with the other

```python
sns.pairplot(df,hue='sex')
```

## **Categoriacal Data Plot**



* Count Plot :- To count and find the total number of different entries in a column

    

```python
df = sns.load_dataset('tips')
df
sns.countplot(x = df['sex'],hue = df['smoker'])
```

* Bar Plots takes two positional argument where one is qualitative label and y is quant label

```python
sns.barplot(x = df['sex'],y = df['total_bill'],estimator=np.sum)
```

* Box plot creates a 5 number summary and take one quant and one qualitative argument

```python
sns.boxplot(x=df['tip'],y=df['day'],data=df,palette='rainbow')
```

```python
sns.violinplot(x=df['day'],y=df['total_bill'],data=df)
```

```python
sns.stripplot(x=df['day'],y=df['total_bill'],data=df)
```

```python
sns.swarmplot(x=df['day'],y=df['total_bill'],data=df)
```
