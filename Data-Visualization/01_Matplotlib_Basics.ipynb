```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

**Matplotlib basics**

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

**Line Plot**

A line plot is used to visualize trends or continuous data.

```python

plt.plot(x,y)
plt.title("Plot")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
```

**subplot in matplot**

Subplots allow multiple plots to be displayed in a single figure.

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.subplot(2,2,4)
plt.subplot(2,2,3)
```

**object oriented way**

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,.4,.4,.4])
axis2.plot(x,y)
```

**Types of Plot**

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,.8,.8])
ax.plot(x,y)
plt.savefig("BasicPlot.png")
```
