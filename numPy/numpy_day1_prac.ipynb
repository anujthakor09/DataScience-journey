Create a dataset of marks for 10 students in 5 subjects using NumPy.

Perform all of the following:

Find the topper.
Find the student with the lowest marks.
Display students scoring above 75% average.
Find subject-wise averages.
Identify the highest score in each subject.
Identify the lowest score in each subject.

```python
marks = np.array([
    [85, 78, 92, 88, 95],
    [67, 74, 70, 72, 80],
    [90, 91, 89, 87, 93],
    [56, 65, 60, 58, 62],
    [77, 81, 79, 84, 86],
    [94, 96, 95, 90, 98],
    [72, 69, 75, 71, 73],
    [88, 85, 90, 86, 89],
    [61, 64, 59, 66, 68],
    [80, 83, 82, 79, 85]
])
```

```python
np.sum(marks,axis=1)
```

```python
np.average(marks,axis=1)
```

```python
np.argmax(np.average(marks,axis=1))
```

```python
np.argmin(np.sum(marks,axis=1))
```

```python
abv_75 = np.average(marks,axis=1)
abv_75
```

```python
abv_75[abv_75 > 75]
```

```python
sub_avg = np.average(marks,axis=0)
```

```python
sub_avg
```

```python
high_scr_sub = np.max(marks,axis=0)
high_scr_sub
```
